# Basic image manipulation using tensorflow

#### Import packages

```python
import numpy as np
import tensorflow as tf

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

#### Load and plot the iris image

```python
filename = "setosa.jpg"
img = mpimg.imread(filename)

plt.imshow(img)
plt.show()
```

#### Rotate the image counter clockwise by transposing the *x* and *y* coordinates

```python
iris = tf.Variable(img, name = 'iris')

model = tf.global_variables_initializer()

with tf.Session() as session:
    iris = tf.transpose(iris, perm=[1, 0, 2])
    session.run(model)
    result = session.run(iris)

plt.imshow(result)
plt.show()
```

#### Rotate the image by 180 degrees

```python
height, width, depth = img.shape

# Create a TensorFlow Variable
iris = tf.Variable(img, name = 'iris')

model = tf.global_variables_initializer()

with tf.Session() as session:
    iris = tf.reverse_sequence(iris, width * [height], 0, batch_dim=1)
    session.run(model)
    result = session.run(iris)

plt.imshow(result)
plt.show()
```

#### Rotate the image clockwise by transposing the *x* and *y* coordinates (using *reverse_sequence* and *transpose*)

```python
height, width, depth = img.shape

# Create a TensorFlow Variable
iris = tf.Variable(img, name = 'iris')

model = tf.global_variables_initializer()

with tf.Session() as session:
    iris = tf.reverse_sequence(iris, width * [height], 0, batch_dim=1)
    iris = tf.transpose(iris, perm=[1, 0, 2])
    session.run(model)
    result = session.run(iris)

plt.imshow(result)
plt.show()
```

#### Horizontal mirror of the image

```python
height, width, depth = img.shape

# Create a TensorFlow Variable
iris = tf.Variable(img, name = 'iris')

model = tf.global_variables_initializer()

with tf.Session() as session:
    iris = tf.reverse_sequence(iris, [width] * height, 1, batch_dim = 0)
    session.run(model)
    result = session.run(iris)

plt.imshow(result)
plt.show()
```

#### Convert to grayscale by colapsing the color dimension of the matrix as a weighted sum

```python
filt = np.array([0.299, 0.587, 0.114])
filt = filt / np.sum(filt)


# Create a TensorFlow Variable
iris = tf.Variable(img, name = 'iris')

model = tf.global_variables_initializer()

with tf.Session() as session:
    iris = img.dot(filt)
    session.run(model)
    result = session.run(iris)

plt.imshow(img.dot(filt), cmap = 'gray')
plt.show()
```
